## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
b=df.groupby('Make')['Cylinders'].mean().plot(kind='bar',title="Cars Visualization")
plt.show()
```

```python
def Cars():
    print(df.Make.unique())
    dc = input ("Enter Cars:")
    r = df[df['Make']==dc]
    print(df.Type.unique())
    dt= input("Enter Type")
    rr = r[r['Type']==dt]
    plt.figure(figsize=(20,6))
    sb.barplot(rr['Model'],rr['MSRP'],label= 'Models')
    #plt.legend(fontsize=13)
    plt.title("%s_Summary"%(dc) ,fontsize=15)
    plt.show()
```
